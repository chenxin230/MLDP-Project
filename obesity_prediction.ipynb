```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import accuracy_score, classification_report
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the data
df = pd.read_csv('Obesity_prediction.csv')
df
```

```python
# Drop the TUE column
df = df.drop(columns=['TUE'])

# Check the DataFrame after dropping the column
print(df.head())
```

```python
# One hot encoding for CAEC, CALC, MTRANS
df = pd.get_dummies(df, columns=['CAEC', 'CALC','MTRANS'])

df.head()
```

```python
# rounding age
df['Age'] = df['Age'].round()

#GENDER (binary)
df['Gender'] = df['Gender'].map({'Male': 1, 'Female': 0})

#SMOKE (binary)
df['SMOKE'] = df['SMOKE'].map({'yes': 1, 'no': 0})

#family_history (binary)
df['family_history'] = df['family_history'].map({'yes': 1, 'no': 0})

#FAVC (binary)
df['FAVC'] = df['FAVC'].map({'yes': 1, 'no': 0})

#SCC
df['SCC'] = df['SCC'].map({'yes': 1, 'no': 0})

df.sample(10)
```

```python
# calculating of the BMI
df['BMI'] = df['Weight'] / (df['Height'] ** 2)

print(df[['Height', 'Weight', 'BMI']].head())
```

```python
# after getting BMI can drop weight and height
df = df.drop(columns=['Weight', 'Height'])

print(df.head())
```

```python
df[df.columns].value_counts()
```

```python
#Obesity column to numeric values
obesity_mapping = {
    'Insufficient_Weight': 0,
    'Normal_Weight': 1,
    'Overweight_Level_I': 2,
    'Overweight_Level_II': 3,
    'Obesity_Type_I': 4,
    'Obesity_Type_II': 5,
    'Obesity_Type_III': 6
}

# apply the mapping to the Obesity column
df['Obesity'] = df['Obesity'].map(obesity_mapping)


df.sample(10)
```

```python
X = df.drop(columns=['Obesity'])  # Drop Obesity
y = df['Obesity']
```

```python
#Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

#Random Forest model
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
```

```python
# importance
importances = rf_model.feature_importances_

# numpy array
feature_importance_array = np.array(importances)

# Sort them
sorted_feature_indexes = feature_importance_array.argsort()[::-1]  # Reverse to get from most to least important

sorted_feature_importance_df = pd.DataFrame({
    'Feature': X.columns[sorted_feature_indexes], 
    'Importance': feature_importance_array[sorted_feature_indexes] 
})

# Print each feature and its importance using a for loop
for index in sorted_feature_indexes:
    print("({}): {:.2f}%".format(X.columns[index], (feature_importance_array[index] * 100)))
```

```python
# Drop column
df = df.drop(columns=['SCC', 'CALC_Always', 'CALC_Frequently','CALC_no','CALC_Sometimes','MTRANS_Motorbike','MTRANS_Bike','MTRANS_Walking','MTRANS_Public_Transportation','MTRANS_Automobile','SMOKE','CAEC_Always','CAEC_no','CAEC_Frequently','CAEC_Sometimes','SCC','FAVC'])
print(df.head())
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

# Create the model
model = RandomForestClassifier(random_state=42)

# Grid search
param_grid = {
   'n_estimators': [50, 100, 200],
   'max_depth': [3, 5, 7, None],
   'min_samples_split': [2, 5, 10],
   'min_samples_leaf': [1, 2, 4]
}

grid_search = GridSearchCV(model, param_grid, cv=5, scoring='accuracy', n_jobs=-1)
grid_search.fit(X_train, y_train)

# Get best model
best_model = grid_search.best_estimator_

# Evaluate on test set
y_pred = best_model.predict(X_test)
print(f"Test Accuracy: {accuracy_score(y_test, y_pred):.2f}")
```

```python
# Retrain test accuracy
threshold = 0.01
important_features = sorted_feature_importance_df[sorted_feature_importance_df['Importance'] > threshold]['Feature']

X_train_important = X_train[important_features]
X_test_important = X_test[important_features]

rf_model = RandomForestClassifier(random_state=42)

# Set up the parameter grid for grid search
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [3, 5, 7, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# grid search
grid_search = GridSearchCV(rf_model, param_grid, cv=5, scoring='accuracy', n_jobs=-1)
grid_search.fit(X_train_important, y_train)

# get the best model from the grid search
best_rf_model = grid_search.best_estimator_

# evaluate the best model
y_pred_retrained = best_rf_model.predict(X_test_important)
print(f"Retrained Test Accuracy: {accuracy_score(y_test, y_pred_retrained):.2f}")
```

```python
# Define parameter grid
from sklearn.tree import DecisionTreeClassifier


# Define parameter grid
param_grid = {
   'max_depth': [3, 5, 7, None],
   'min_samples_split': [2, 5, 10],
   'min_samples_leaf': [1, 2, 4],
   'criterion': ['gini', 'entropy']
}

# Create GridSearchCV object
grid_search = GridSearchCV(
   estimator=DecisionTreeClassifier(random_state=42),
   param_grid=param_grid,
   cv=5,
   scoring='accuracy',
   n_jobs=-1
)

# Perform grid search
grid_search.fit(X_train, y_train)

# Get best model
dt_model = grid_search.best_estimator_

# Make predictions
y_pred = dt_model.predict(X_test)

print("Test Accuracy:", accuracy_score(y_test, y_pred))
```

```python
threshold = 0.01 
important_features = sorted_feature_importance_df[sorted_feature_importance_df['Importance'] > threshold]['Feature']


X_train_important = X_train[important_features]
X_test_important = X_test[important_features]

param_grid = {
   'max_depth': [3, 5, 7, None],
   'min_samples_split': [2, 5, 10],
   'min_samples_leaf': [1, 2, 4],
   'criterion': ['gini', 'entropy']
}

grid_search = GridSearchCV(
   estimator=DecisionTreeClassifier(random_state=42),
   param_grid=param_grid,
   cv=5,
   scoring='accuracy',
   n_jobs=-1
)

grid_search.fit(X_train_important, y_train)

# Get best model from grid search
dt_model_retrained = grid_search.best_estimator_

y_pred_retrained = dt_model_retrained.predict(X_test_important)

# Print retrained accuracy
print("Retrained Test Accuracy:", accuracy_score(y_test, y_pred_retrained))
```

```python
from sklearn.metrics import confusion_matrix, classification_report

# Generate and print the confusion matrix
print("Confusion matrix:")
conf_matrix = confusion_matrix(y_test, y_pred_retrained)
print(conf_matrix)

# Print the classification report
print("Classification report:")
print(classification_report(y_test, y_pred_retrained))
```

```python
# Save the best Decision Tree retrained model
import joblib

joblib.dump(dt_model_retrained, 'obesity_decision_tree_retrained_model.pkl')

# Optional: Verify the saved retrained model
loaded_retrained_model = joblib.load('obesity_decision_tree_retrained_model.pkl')

# Confirm model performance remains the same on the important features
loaded_retrained_predictions = loaded_retrained_model.predict(X_test_important)
print("Retrained Model Accuracy:", accuracy_score(y_test, loaded_retrained_predictions))
```
